```python
import pandas as pd 
```

```python
adver_data=pd.read_csv("advertising.csv")
print(adver_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = adver_data[['TV', 'Radio', 'Newspaper']].values
y = adver_data['Sales'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, feature in enumerate(['TV', 'Radio', 'Newspaper']):
    axes[i].scatter(X_test[:, i], y_test, label='Actual Sales', alpha=0.7)
    axes[i].scatter(X_test[:, i], y_pred, color='red', marker='x', label='Predicted Sales')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Sales')
    axes[i].legend()

plt.suptitle('Sales Prediction using Linear Regression', y=1.02)
plt.show()
```

```python
new_advertising_expenditure = np.array([[100, 50, 25]])
predicted_sales = model.predict(new_advertising_expenditure)
print(f'Predicted Sales for New Advertising Expenditure: {predicted_sales[0]}')
```
